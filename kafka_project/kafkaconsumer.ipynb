```python
consumer = KafkaConsumer(
    'sampletopic',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='latest',
    value_deserializer=lambda m: json.loads(m.decode('utf-8'))
)

print("Listening for messages...")
```

```python
csv_file = "processed_records.csv"
```

```python
fieldnames = ["age","job","marital","balance","housing","loan","credit_card_eligible"]
```

```python

# for msg in consumer:
#     data = msg.value
    
#     selected = {col: data[col] for col in required_cols}

#     print("Selected:",  selected)
```

```python
required_cols = [
    "age", "job", "marital", "balance", "housing", "loan"
]
```

```python
from numpy import record


for message in consumer:
    data = message.value
    selected = {col: data[col] for col in required_cols}
    job = selected["job"]
    balance = selected["balance"]
    
    if job.lower() != "unemployed" and balance > 500:
        credit_eligible = "yes"
    else:
        credit_eligible = "no"

    # ---- Add new column ----
    selected["credit_card_eligible"] = credit_eligible
    

    with open(csv_file, "a", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writerow(selected)

    print("Saved:", selected)

    # save_to_cassandra(selected)
```

```python
from cassandra.cluster import Cluster

cluster = Cluster(['localhost'])
session = cluster.connect('finance')

insert_query = session.prepare("""
    INSERT INTO customers (age, job, marital, balance, housing, loan, credit_card_eligible)
    VALUES (?, ?, ?, ?, ?, ?, ?)
""")

def save_to_cassandra(record):
    session.execute(insert_query, (
        record["age"],
        record["job"],
        record["marital"],
        record["balance"],
        record["housing"],
        record["loan"],
        record["credit_card_eligible"]
    ))
    print("Inserted into Cassandra:", record)
```
