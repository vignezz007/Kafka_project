```python
import pandas as pd
import time
from kafka import KafkaProducer
import json
import os
```

```python
source_df = pd.read_csv('E:\\kafka_project\\bank.csv', sep=';')
target_df = pd.DataFrame(columns=source_df.columns)
```

```python
producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)
```

```python
TOPIC = "sampletopic"
```

```python
last_data = None  # track last sent row

print("Kafka Producer watching for new rows...\n")

while True:

    if os.path.exists("new_row.json"):
        with open("new_row.json", "r") as f:
            data = json.load(f)

        # Send only if new
        if data != last_data:
            producer.send(TOPIC, data)
            producer.flush()
            last_data = data
            print("Sent to Kafka:", data)

    time.sleep(1)   # check every second
```
