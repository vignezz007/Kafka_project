```python
import time
import json
```

```python
datapath = "E:\\kafka_project\\bank.csv"
```

```python
source_df = pd.read_csv('E:\\kafka_project\\bank.csv', sep=';')
```

```python
source_df.head(5)
```

```python
source_df.info()
```

```python
target_df = pd.DataFrame(columns=source_df.columns)
```

```python
for idx, row in source_df.iterrows():

    # Append to target df
    target_df = pd.concat([target_df, pd.DataFrame([row])], ignore_index=True)

    # Save newly added row so producer can read it
    with open("new_row.json", "w") as f:
        json.dump(row.to_dict(), f)

    print("Appended row:", row.to_dict())
    time.sleep(5)

print("Done appending rows.")
```
